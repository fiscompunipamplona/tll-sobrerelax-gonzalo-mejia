Considere la ecuación $x = 1 - e^{-cx}$, donde $c$ es un parámetro conocido y $x$ es desconocido. Esta ecuación surge en una variedad de situaciones, incluida la física de los procesos de contacto, los modelos matemáticos de epidemias y la teoría de los gráficos aleatorios.

a) Escriba un programa para resolver esta ecuación para $x$ usando el método de relajación para el caso $c = 2$. Calcule su solución con una precisión de al menos $10^{- 6}$.

b) Modifique su programa para calcular la solución para valores de c de 0 a 3 en pasos de 0.01 y haga un gráfico de x en función de c. Debería ver una transición clara de un régimen en el que x = 0 a un régimen de x distinto de cero. Este es otro ejemplo de una transición de fase. En física, esta transición se conoce como transición de percolación; en epidemiología es el umbral epidémico.

```python
from scipy import exp, linspace
from pylab import plot, show, xlabel, ylabel
```

```python
# Constants
precision = 10e-6

def g(c):
    interaciones = 1
    def f(m):
        return 1 - exp(- c * m)

    def error(m1, m2):
        return (m1 - m2) / (1 - 1 / (c * exp(-c * m1)))

    m1 = 0.5 # valor inicial
    m2 = f(m1)
    while(abs(error(m1, m2)) > precision):
        m1, m2 = m2, f(m2)
        interaciones += 1
    print('numero de interaciones = ', interaciones)
    return m2

c = linspace(0, 3, 300)
x = list(map(g, c))
```

```python
plot(c, x, 'r')
xlabel('c')
ylabel('x')
show()
```
